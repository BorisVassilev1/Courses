```python
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("train.csv")
data = data.drop(columns = "Id")

for col in data.columns:
    if data.dtypes[col] == 'object':
        data[col] = data[col].fillna("---")
        data[col]
    else:
        data[col] = data[col].fillna(data[col].mean())
data
```

```python
def pltcol(col, width = 0.9):
    plt.figure(figsize=(20,7))
    plt.bar(train_init[col].unique(), list((train_init[train_init[col] == x]["SalePrice"].mean() for x in train_init[col].unique())), width = width)
    plt.show()
```

```python
def dummies(col):
    data[sorted(("is" + col + "_" + x) for x in data[col].unique())] = pd.get_dummies(data[col], prefix = "is" + col)
```

```python
for col in data.columns:
    if data.dtypes[col] == 'object':
        dummies(col)
        if ("is" + col + "_" + "---") in data.columns:
            data = data.drop(columns = ("is" + col + "_" + "---"))
        data = data.drop(columns = col)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

X_init = data.drop(columns = ["SalePrice"])
Y_init = data.SalePrice

X_train, X_val, Y_train, Y_val = train_test_split(X_init, 
                                                  Y_init, 
                                                  test_size=0.1,  
                                                  random_state=42)



lin_reg = LinearRegression()

lin_reg.fit(X_train, Y_train)

lin_reg.score(X_val, Y_val)
```
